```python
steel = pd.read_csv("steel.csv")
print(steel.head())
```

```python
steel.index = pd.to_datetime(steel['Date'], format='%Y-%m-%d')
steel.index
```

```python
del steel['Date']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.plot(steel.index, steel['STEEL'], )
```

```python
train = steel[steel.index < pd.to_datetime("2019-11-01", format='%Y-%m-%d')]
test = steel[steel.index > pd.to_datetime("2019-11-01", format='%Y-%m-%d')]

plt.plot(train, color = "black")
plt.plot(test, color = "red")
plt.ylabel('BTC Price')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.title("Train/Test split for BTC Data")
plt.show()
```
