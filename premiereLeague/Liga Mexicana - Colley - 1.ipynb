```python
#data_file = './data/FMF_TA_2021.csv'
data_file = './data/lmf-ac-2021-22.csv'
df = pd.read_csv(data_file, index_col=0)
df.head()
```

```python
print(df.columns)
```

```python
df.index = df.columns
rows = []
for i in df.index:
    for c in df.columns:
        if i == c: continue
        score = df.loc[i, c]
        if score == '—': continue
        ssplit = score.split('–')
        rows.append([i, c, ssplit[0], ssplit[1]])
df = pd.DataFrame(rows, columns = ['home', 'away', 'home_score', 'away_score'])
df.head()
```

```python
df.dtypes
```

```python
df.home_score = df['home_score'].astype('int')
df.away_score = df['away_score'].astype('int')
```

```python
colley = pb.ratings.Colley(df["home_score"], df["away_score"], df["home"], df["away"])
colley.get_ratings()
```

```python
colley = pb.ratings.Colley(df["home_score"], df["away_score"], df["home"], df["away"], include_draws=False)
colley.get_ratings()
```
