```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pingouin as pg
```

```python
df = pd.read_csv('./data/f1-2022-1-13.csv')
df
```

```python
# Datos
# ==============================================================================
valores_A = np.array([0.0,25.0,0.0,34.0,26.0,25.0,15.0,25.0,25.0,6.0,27.0,25.0,25.0])
valores_B = np.array([26.0,19.0,26.0,15.0,18.0,0.0,12.0,0.0,10.0,12.0,32.0,0.0,8.0])
muestra = np.repeat(['A', 'B'], repeats=13)
datos = pd.DataFrame({
            'valor': np.concatenate([valores_A, valores_B]),
            'muestra': muestra
})
datos
```

```python
# Test de homocedasticidad
# ==============================================================================
pg.homoscedasticity(data=datos, dv='valor', group='muestra', method='levene')
```

```python
# U-test (p-value y tamaño de efecto)
# ==============================================================================
pg.mwu(x=valores_A, y=valores_B, alternative='two-sided')
```
