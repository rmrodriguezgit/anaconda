```python
#import cv2, numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("../F1-Data-Extractor/images/IMG_0072.jpeg")
#Displaying image using plt.imshow() method
b,g,r = cv2.split(img)
rgb_img = cv2.merge([r,g,b])
plt.imshow(rgb_img)
```

```python
imageo = rgb_img
```

```python
## Tamao de la imagen
```

```python
# Extracting the height and width of an image
h, w = imageo.shape[:2]
# Displaying the height and width
print("Height = {},  Width = {}".format(h, w))
```

```python
new_height = 1920
new_width = 1440

# dsize
dsize = (new_width, new_height)

# resize image
image = cv2.resize(imageo, dsize, interpolation = cv2.INTER_AREA)
```

```python
# Extracting the height and width of an image
h, w = image.shape[:2]
# Displaying the height and width
print("Height = {},  Width = {}".format(h, w))
```

```python
# Extracting RGB values. 
# Here we have randomly chosen a pixel
# by passing in 100, 100 for height and width.
(B, G, R) = image[1649, 1114]
  
# Displaying the pixel values
print("R = {}, G = {}, B = {}".format(R, G, B))
  
# We can also pass the channel to extract 
# the value for a specific channel
B = image[1649, 1114, 2]
print("B = {}".format(B))
```

```python
image.item(1649,1114,2)
```

```python
# We will calculate the region of interest 
# by slicing the pixels of the image
# Coordinates [y1:y2 , x1:x2]
 
roi_driver = image[1020:1048,286:362]
roi_team = image[1050:1078,286:450]
roi_tirelap = image[1157:1193,1337:1369]
roi_tire = image[1165:1182,1381:1401]
roi_s1 = image[1269:1294,1204:1295]
roi_s2 = image[1325:1356,1204:1295]
roi_s3 = image[1387:1421,1204:1295]
roi_best = image[1442:1472,1204:1295]
roi_velocity = image[1412:1495,417:619]
roi_rpm = image[1542:1586,431:602]
roi_gear = image[1714:1753,537:576]

#plt.imshow(roi_tire)
plt.imshow(roi_velocity)
#plt.imshow(roi_gear)
```

```python
# Tesseract
custom_config = r'--oem 1 --psm 6'
driver = pytesseract.image_to_string(roi_driver, config=custom_config)
team = pytesseract.image_to_string(roi_team, config=custom_config)
tirelap = pytesseract.image_to_string(roi_tirelap, config=custom_config)
custom_config = r'--oem 0 --psm 6'
tire = pytesseract.image_to_string(roi_tire, config=custom_config)
custom_config = r'--oem 1 --psm 6'
s1 = pytesseract.image_to_string(roi_s1, config=custom_config)
s2 = pytesseract.image_to_string(roi_s2, config=custom_config)
s3 = pytesseract.image_to_string(roi_s3, config=custom_config)
best = pytesseract.image_to_string(roi_best, config=custom_config)
velocity = pytesseract.image_to_string(roi_velocity, config=custom_config)
rpm = pytesseract.image_to_string(roi_rpm, config=custom_config)
gear = pytesseract.image_to_string(roi_gear, config=custom_config)

print('Extraccin de informacin para telemetra F1')
plt.imshow(rgb_img)
plt.title('Imagen Original')
plt.show()

print('Presentacin de la informacin como texto \n')
print(f'piloto = {driver}')
print(f'equipo = {team}')
print(f'tirelap = {tirelap}')
print(f'tire = {tire}')
print(f's1 = {s1}')
print(f's2 = {s2}')
print(f's3 = {s3}')
print(f'best = {best}')
print(f'velocity = {velocity}')
print(f'rpm = {rpm}')
print(f'gear = {gear}')

```
