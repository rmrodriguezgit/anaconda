```python
#import cv2, numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("../F1-Data-Extractor/images/f1-22-cp.png")
#Displaying image using plt.imshow() method
b,g,r = cv2.split(img)
rgb_img = cv2.merge([r,g,b])
plt.imshow(rgb_img)
```

```python
imageo = rgb_img
```

```python
## Tamao de la imagen
```

```python
# Extracting the height and width of an image
h, w = imageo.shape[:2]
# Displaying the height and width
print("Height = {},  Width = {}".format(h, w))
```

```python
new_height = 720
new_width = 960
# dsize
dsize = (new_width, new_height)

# resize image
image = cv2.resize(imageo, dsize, interpolation = cv2.INTER_AREA)
```

```python
# Extracting the height and width of an image
h, w = image.shape[:2]
# Displaying the height and width
print("Height = {},  Width = {}".format(h, w))
```

```python
# Extracting RGB values. 
# Here we have randomly chosen a pixel
# by passing in 100, 100 for height and width.
(B, G, R) = image[100, 100]
  
# Displaying the pixel values
print("R = {}, G = {}, B = {}".format(R, G, B))
  
# We can also pass the channel to extract 
# the value for a specific channel
B = image[100, 100, 2]
print("B = {}".format(B))
```

```python
image.item(100,100,2)
```

```python
# We will calculate the region of interest 
# by slicing the pixels of the image
# Coordinates [y1:y2 , x1:x2]
 
#		
roi_laptime = image[25:46, 24:153]
roi_s1 = image[69:85, 884:942]
roi_s2 = image[98:114, 884:942]
roi_s3 = image[123:140, 884:942]
roi_gear = image[652:691, 705 : 751]
roi_velocity = image[652: 679 , 828 : 893 ]
roi_rpm = image[683:695  , 875 : 912]

#plt.imshow(roi_tire)
#plt.imshow(roi_best)
plt.imshow(roi_rpm)
```

```python
fig = plt.figure()
ax = fig.add_subplot(2, 2, 1)
ax.imshow(roi_laptime)
ax.axis('off')
ax = fig.add_subplot(2, 2, 2)
ax.imshow(roi_s3)
ax.axis('off')
ax = fig.add_subplot(2, 2, 3)
ax.imshow(roi_rpm)
ax.axis('off')
ax = fig.add_subplot(2, 2, 4)
ax.imshow(roi_gear)
ax.axis('off')
```

```python
# Tesseract
custom_config = r'--oem 2 --psm 6'
laptime = pytesseract.image_to_string(roi_laptime, config=custom_config)
custom_config = r'--oem 1 --psm 6'
s1 = pytesseract.image_to_string(roi_s1, config=custom_config)
s2 = pytesseract.image_to_string(roi_s2, config=custom_config)
s3 = pytesseract.image_to_string(roi_s3, config=custom_config)
custom_config = r'--oem 2 --psm 6'
gear = pytesseract.image_to_string(roi_gear, config=custom_config)
custom_config = r'--oem 1 --psm 6'
velocity = pytesseract.image_to_string(roi_velocity, config=custom_config)
custom_config = r'--oem 2 --psm 6'
rpm = pytesseract.image_to_string(roi_rpm, config=custom_config)



print('Extraccin de informacin para telemetra F1')
plt.imshow(rgb_img)
plt.title('Imagen Original')
plt.show()

print('Presentacin de la informacin como texto \n')
print(f'laptime = {laptime}')
print(f's1 = {s1}')
print(f's2 = {s2}')
print(f's3 = {s3}')
print(f'velocity = {velocity}')
print(f'rpm = {rpm}')
print(f'gear = {gear}')

```
