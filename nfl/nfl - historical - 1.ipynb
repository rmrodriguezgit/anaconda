```python
data_file = './data/spreadspoke_scores.csv'
df = pd.read_csv(data_file, index_col=0)
df.head()
```

```python
# 0 Jacksonville Jaguars
# 1 Indianapolis Colts
# 2 New Orleans Saints
# 3 Cincinnati Bengals
# 4 Minnesota Vikings
# 5 Miami Dolphins
# 6 New England Patriots
# 7 New York Jets
# 8 Green Bay Packers
# 9 Baltimore Ravens
# 10 New York Giants
equipo=['Jacksonville Jaguars','Indianapolis Colts','New Orleans Saints','Cincinnati Bengals',
        'Minnesota Vikings','Miami Dolphins','New England Patriots','New York Jets','Green Bay Packers',
       'Baltimore Ravens','New York Giants']
```

```python
E1 = equipo[3]
E2 = equipo[2]
```

```python
df2 = df.loc[(df['team_home'] == E1) & (df['team_away'] == E2)]
df2
```

```python
df2['ganador'] = np.where(df2['score_home']>df2['score_away'], 'L','V')
df2['diferencial'] = df2['score_home'] - df2['score_away']
df2
```

```python
df2.loc[df2['ganador']=='L']
```

```python
l = sum(df2.ganador == 'L')
v = sum(df2.ganador == 'V')
print (f'{E1} - local: {l} \n  {E2} - visita: {v}')
```
